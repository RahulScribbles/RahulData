# 2 Calculate the average profit margin for each product.

```python
sales_data['average profit'] = (sales_data['Profit'] - sales_data['Price']) / sales_data['Profit'] * 100
sales_data
```

```python
groupby = sales_data.groupby('ProductID')
```
