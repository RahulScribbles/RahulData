# 1 Plotting x and y points using Matplotlib Plotting

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
#Draw a line in a diagram from position (1, 3) to position (8, 10):
```

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])
```

```python
plt.plot(xpoints, ypoints)
plt.show()
```

```python


#The x-axis is the horizontal axis.

#The y-axis is the vertical axis.
```

# Format Strings fmt

You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this 


syntax:
marker|line|color

```python
#Dotted Line 

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints, 'o:m')
plt.show()
```

# Line Reference



'-'   solid Line

':'   dotted line

'--'  Dashed line

'-.'  Dashed/ dotted line

```python
#Dashed/ Dotted Line  '-.'

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints, 'o-.g')
plt.show()
```

# Color reference

'r'   Red

'g'  Green

'b'  Blue

'c'  Cyan

'm'  Magenta

'y'  Yellow

'k'  Black

'w'  White

# Marker Size

You can use the keyword argument 'markersize' or the shorter version, 'ms' to set the size of the markers:

```python
#Example

#Set the size of markers to 20
```

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints, marker = 'o', ms=20)
plt.show()
```

# Marker Color


You can use the keyword argument 'markeredgecolor' or 
the shorter 'mec' to set the color of the edge of the markers:

```python
#Example

#Set the EDGE color To red

xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints, marker = 'o', ms=20, mec = 'r')
plt.show()
```

# You can use the keyword argument 'markerfacecolor' or the shorter 'mfc' to set the color inside the edge of the markers:

```python
#Example

#Set the Face color to red
```

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints, marker = 'o', ms=20, mfc = 'r')
plt.show()
```

# Use both the 'mec' and 'mfc' arguments to color the entire marker

```python
#example

#set the color of both the edge and face to red
```

```python
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])

plt.plot(xpoints, ypoints, marker = 'o', ms=20, mec = 'r', mfc = 'r')
plt.show()
```
