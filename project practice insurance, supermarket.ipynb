# Project-1 (Health Insurance Premium Prediction)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("D:\\Ed_Bytes Data Analytics lecture\\Python lecture\\project\\L 29 data file insurance.csv")
```

# So lets have look at distribution of people who smoke or not

```python
import plotly.express as px
```

```python
figure = px.histogram(df ,
                     x = 'sex',
                     color = 'smoker',
                     title = 'number of smokers')
figure.show()
```

# So , now we will replace the value of "sex" and "smoker columns with 0 and 1 as both these columns containing string values .

```python
df['sex'] = df['sex'].map({'female':0, 'male':1})
```

```python
df['smoker'] = df['smoker'].map({'no':0 , 'yes':1})
```

```python
x = np.array(df[['age' , 'sex' , 'bmi' , 'smoker']])
y = np.array(df['charges'])
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
x_train.shape
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train , y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test , y_pred)
```

# Project-2(Product-demand Prediction)

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("D:\\Ed_Bytes Data Analytics lecture\\Python lecture\\project\\L29 Data file supermarket.csv")
```

```python
df
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fig = px.scatter(df,
                x = "Unit price",
                y = "Total",
                size = "Quantity")
fig.show()
```

```python
df.isnull().sum()
```

```python
x = df[['Unit price' , 'Quantity']]
#x
```

```python
y = df['Total']
#y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y, test_size = 0.2 , random_state = 42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(x_train , y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
r2_score(y_test , y_pred)
```
